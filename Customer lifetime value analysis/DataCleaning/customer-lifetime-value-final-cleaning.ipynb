# Customer Revenue sum updation

```python
df_new['Customer_Revenue_Sum'] = (
    df_new[df_new['Is_Valid_Sale'] == 1]
    .groupby('Customer ID')['Revenue']
    .transform('sum')
)

df_new['Customer_Revenue_Sum'] = (
    df_new['Customer_Revenue_Sum']
    .fillna(0)
)
```

```python
print(df_new['Customer_Revenue_Sum'].head())
```

# Recency_Days updation

```python
df_new['InvoiceDate'] = pd.to_datetime(df_new['InvoiceDate'])

analysis_date = df_new['InvoiceDate'].max()

df_new['Recency_Days'] = (
    analysis_date
    - df_new[df_new['Is_Valid_Sale'] == 1]
        .groupby('Customer ID')['InvoiceDate']
        .transform('max')
).dt.days

df_new['Recency_Days'] = df_new['Recency_Days'].fillna(0).astype(int)
print(df_new[['Customer ID','Recency_Days']].drop_duplicates().head(10))
```

# customer_invoice_count updation

```python
df_new['Customer_Invoice_Count'] = (
    df_new[df_new['Is_Valid_Sale'] == 1]
        .groupby('Customer ID')['Invoice']
        .transform('nunique')
)

df_new['Customer_Invoice_Count'] = df_new['Customer_Invoice_Count'].fillna(0).astype(int)

```

# Monetary_Value updation

```python
df_new['Revenue'] = pd.to_numeric(df_new['Revenue'], errors='coerce').fillna(0)

df_new['Total_Valid_Revenue'] = (
    df_new[df_new['Is_Valid_Sale'] == 1]
        .groupby('Customer ID')['Revenue']
        .transform('sum')
)

df_new['Monetary_Value'] = df_new['Total_Valid_Revenue'] / df_new['Customer_Invoice_Count']

df_new['Monetary_Value'] = (
    df_new['Monetary_Value']
        .replace([float('inf'), -float('inf')], 0)
        .fillna(0)
        .round(2)
)

df_new.drop(columns=['Total_Valid_Revenue'], inplace=True)

print(
    df_new[['Customer ID', 'Customer_Invoice_Count', 'Monetary_Value']]
    .drop_duplicates()
    .head(10)
)
```

# Avg_Order_Value updation

```python
df_new['Revenue'] = pd.to_numeric(df_new['Revenue'], errors='coerce').fillna(0)
df_new['Invoice'] = df_new['Invoice'].astype(str)

valid_sales = df_new[df_new['Is_Valid_Sale'] == 1].copy()
valid_sales['Invoice'] = valid_sales['Invoice'].astype(str)

invoice_revenue = valid_sales.groupby('Invoice', as_index=False)['Revenue'].sum().rename(columns={'Revenue':'Avg_Order_Value'})

df_new = df_new.merge(invoice_revenue, on='Invoice', how='left')

df_new['Avg_Order_Value'] = df_new['Avg_Order_Value'].fillna(0).round(2)

print(df_new[['Invoice','Avg_Order_Value']].drop_duplicates().head(10))
```

# Financial_impact updation

```python
df_new['Financial_Impact'] = df_new.apply(lambda x: x['Quantity'] * x['Price'] if x['Is_Valid_Sale'] == 0 else 0, axis=1)

print(df_new[['Invoice','Customer ID','Quantity','Is_Valid_Sale','Non_Sale_Quantity','Financial_Impact']].head(10))
```
