# Dropping columns to recreate those with better logic

# Invoice_Type Column Creation

```python
df_retail['Invoice_Type'] = df_retail['Invoice'].astype(str).str.startswith("C")
df_retail['Invoice_Type']= df_retail['Invoice_Type'].replace({True:"Return", False:"Sale"})
df_retail.loc[df_retail['Quantity']<0,'Invoice_Type']='Return'

df_retail.loc[(df_retail['Invoice'].astype(str).str.startswith("C"))&(df_retail['Quantity']>0),'Invoice_Type']="Check Manual"

```

# Is_Valid_Sale column creation

```python
df_retail['Is_Valid_Sale']=0
valid_sale_condition=((df_retail['Invoice_Type']=='Sale')&
                        (df_retail['Quantity']>0)&
                        (df_retail['Price'] > 0)&
                      (df_retail['Customer ID'].notnull()) 
                     )
df_retail.loc[valid_sale_condition,'Is_Valid_Sale']=1

                                         
```

```python
print(df_retail['Is_Valid_Sale'].value_counts())
```

```python
df_retail['Price'] = pd.to_numeric(df_retail['Price'], errors='coerce').fillna(0)
df_retail['Quantity'] = pd.to_numeric(df_retail['Quantity'], errors='coerce').fillna(0)
df_retail['TotalAmount'] = df_retail['Quantity'] * df_retail['Price']
```

```python
# If both are negative, flip them
df_retail.loc[(df_retail['Quantity'] < 0) & (df_retail['Price'] < 0), ['Quantity', 'Price']] *= -1

df_retail['TotalAmount_New'] = df_retail['Quantity'] * df_retail['Price']
df_retail[['Quantity','Price','TotalAmount','TotalAmount_New']].head(10)
```

# Reason_Category column creation

```python
desc= df_retail["Description"].astype(str).str.lower()

def get_reason(desc,qty,invoice_type):

    if qty<0 and invoice_type =="Return":
        return "Return"
    
    fraud_keywords=['false order', 'check', 'found']
    if any(k in desc for k in fraud_keywords):
        return "Fraud / False Order"
    
    manual_keywords = ['manual', 'discount', 'bank charge', 'amazon fee']
    if any(k in desc for k in manual_keywords):
        return "Manual Adjustment"
    
    Damaged_keuwords=['damag', 'damage', 'loss', 'lost']
    if any (k in desc for k in Damaged_keuwords):
        return "Damaged / Missing Item"
    
    if 'postage' in desc:
        return "Postage / Shipping Fee"
    
    if 'sample' in desc:
        return "Postage / Shipping Fee"
    
    if desc in ['?', '??','???','###'] or len(desc) < 3:
        return "Unknown / Miscellaneous"
    
    return "Valid Sale"

df_retail["Reason_Category"]=df_retail.apply(lambda row: get_reason(str(row['Description']).lower(), row["Quantity"], row["Invoice_Type"]), axis=1)
       



```

# Lost_Quantity column creation

# Financial_Impact column creation

# Customer_Type creation

```python
df_retail['Customer_Type'] = df_retail['Customer ID'].apply(lambda x: 'Guest' if x.strip().lower() == 'guest customer' else 'Customer')

# check
df_retail[['Customer ID','Customer_Type']].drop_duplicates().head(10)
```

# Customer_Invoice_Count column creation

```python
# Step 1: valid sales filter 
valid_sales = df_retail[df_retail['Is_Valid_Sale'] == 1]

df_customer_invoice = valid_sales.groupby('Customer ID')['Invoice'].nunique().reset_index()
df_customer_invoice.rename(columns={'Invoice': 'Customer_Invoice_Count'}, inplace=True)

df_retail = df_retail.merge(df_customer_invoice, on='Customer ID', how='left')

df_retail['Customer_Invoice_Count'] = df_retail['Customer_Invoice_Count'].fillna(0).astype(int)

# Check 
df_retail[['Customer ID','Customer_Invoice_Count']].drop_duplicates().head(10)

```

# Customer_Quantity_Sum creation

```python
# Step 1: valid sales filter
valid_sales = df_retail[df_retail['Is_Valid_Sale'] == 1]

df_customer_qty = valid_sales.groupby('Customer ID')['Quantity'].sum().reset_index()
df_customer_qty.rename(columns={'Quantity': 'Customer_Quantity_Sum'}, inplace=True)

df_retail = df_retail.merge(df_customer_qty, on='Customer ID', how='left')

df_retail['Customer_Quantity_Sum'] = df_retail['Customer_Quantity_Sum'].fillna(0).astype(int)

# Check
df_retail[['Customer ID','Customer_Quantity_Sum']].drop_duplicates().head(10)

```

# Customer_Revenue_Sum

```python
# Step 1: valid sales filter
valid_sales = df_retail[df_retail['Is_Valid_Sale'] == 1]

df_customer_revenue = valid_sales.groupby('Customer ID')['Revenue'].sum().reset_index()
df_customer_revenue.rename(columns={'Revenue': 'Customer_Revenue_Sum'}, inplace=True)

df_retail = df_retail.merge(df_customer_revenue, on='Customer ID', how='left')

df_retail['Customer_Revenue_Sum'] = df_retail['Customer_Revenue_Sum'].fillna(0)

# Check
df_retail[['Customer ID','Customer_Revenue_Sum']].drop_duplicates().head(10)
```

# Recency_Days

```python

# Step 1: valid sales only filter
valid_sales = df_retail[df_retail['Is_Valid_Sale'] == 1]


last_purchase = valid_sales.groupby('Customer ID')['InvoiceDate'].max().reset_index()
last_purchase.rename(columns={'InvoiceDate':'Last_Purchase_Date'}, inplace=True)

analysis_date = df_retail['InvoiceDate'].max()

last_purchase['Recency_Days'] = (analysis_date - last_purchase['Last_Purchase_Date']).dt.days

df_retail = df_retail.merge(last_purchase[['Customer ID','Recency_Days']], on='Customer ID', how='left')

df_retail['Recency_Days'] = df_retail['Recency_Days'].fillna(0).astype(int)

# Check
df_retail[['Customer ID','Recency_Days']].drop_duplicates().head(10)
```

# Monetary_Value

```python
# Step 1: valid sales only
valid_sales = df_retail[df_retail['Is_Valid_Sale'] == 1]

customer_revenue = valid_sales.groupby('Customer ID')['Revenue'].sum().reset_index()
customer_revenue.rename(columns={'Revenue':'Total_Revenue'}, inplace=True)

customer_invoice_count = valid_sales.groupby('Customer ID')['Invoice'].nunique().reset_index()
customer_invoice_count.rename(columns={'Invoice':'Invoice_Count'}, inplace=True)

customer_monetary = pd.merge(customer_revenue, customer_invoice_count, on='Customer ID', how='left')
customer_monetary['Monetary_Value'] = customer_monetary['Total_Revenue'] / customer_monetary['Invoice_Count']

df_retail = df_retail.merge(customer_monetary[['Customer ID','Monetary_Value']], on='Customer ID', how='left')

df_retail['Monetary_Value'] = df_retail['Monetary_Value'].fillna(0)

# Check
df_retail[['Customer ID','Monetary_Value']].drop_duplicates().head(10)
```

# Avg_Order_Value (Basically, Invoice-wise total Revenue assign )

```python
# Step 1: valid sales only
valid_sales = df_retail[df_retail['Is_Valid_Sale'] == 1]

invoice_revenue = valid_sales.groupby('Invoice')['Revenue'].sum().reset_index()
invoice_revenue.rename(columns={'Revenue':'Avg_Order_Value'}, inplace=True)

df_retail = df_retail.merge(invoice_revenue, on='Invoice', how='left')

df_retail['Avg_Order_Value'] = df_retail['Avg_Order_Value'].fillna(0)

# Check
df_retail[['Invoice','Avg_Order_Value']].drop_duplicates().head(10)
```
